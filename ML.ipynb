```python
#basic libraries
import pandas as pd
import numpy as np

#scikitlearn libraries for K-Means Clustering Algorithm implementation
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import pairwise_distances_argmin

X, y_true = make_blobs( n_samples=500, centers=9, random_state=0 )
plt.scatter(X[:,0], X[:, 1])
plt.title("Generated Data")
plt.show()

model = KMeans(n_clusters=9)
model.fit(X)
y_pred = model.predict(X)
print(y_pred)

centers = model.cluster_centers_

plt.scatter( X[:,0], X[:,1] )
plt.scatter( centers[:,0], centers[:,1] )
plt.show()



def find_clusters(X, n_clusters, rseed=2):

    rng = np.random.RandomState(rseed)

    i = rng.permutation(X.shape[0])[:n_clusters]

    centers = X[i]

    while True:
        labels = pairwise_distances_argmin(X, centers)
        new_centers = np.array([X[labels==i].mean(0) for i in range(n_clusters)])
    
        if np.all( centers == new_centers ):
            break

    centers = new_centers
    return centers, labels
```

```python
#DecisionTree Implementation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv("/home/aditya/Downloads/Student_Performance.csv")

le = LabelEncoder()
data['Subject'] = le.fit_transform(data['Subject'])      # Maths = 1, Physics = 2...
data['Tuition'] = le.fit_transform(data['Tuition'])      # Yes = 1, No = 0
data['Result'] = le.fit_transform(data['Result'])        # Yes = 1, No = 0



X = data.values[:, 0:3]
y = data.values[:, 3]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

clf_entropy = DecisionTreeClassifier( criterion="entropy", random_state=100, max_depth=3, min_samples_leaf=5 )

clf_entropy.fit(X_train, y_train)

y_pred = clf_entropy.predict(X_test)

print(y_pred)
print(y_test)

print("Accuracy Score: ", accuracy_score(y_test, y_pred)*100)

plt.figure(figsize=(12,8))
tree.plot_tree(clf_entropy, filled=True, feature_names=["Attendance", "Marks", "Assignments"])
plt.show()

```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("/home/aditya/Downloads/Student_Performance.csv")

# Create a new column 'Performance' based on Test_Score
data['Performance'] = ['High' if score >= 80 else 'Low' for score in data['Test_Score']]

# Drop unneeded columns
data = data.drop(['Student', 'Test_Score'], axis=1)

# Convert categorical data to numerical using LabelEncoder
le_subject = LabelEncoder()
le_tuition = LabelEncoder()
le_performance = LabelEncoder()

data['Subject'] = le_subject.fit_transform(data['Subject'])
data['Tuition'] = le_tuition.fit_transform(data['Tuition'])
data['Performance'] = le_performance.fit_transform(data['Performance'])

# Features and target
X = data[['Subject', 'Hours_Studied', 'Tuition']]
y = data['Performance']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Create and train the model
model = DecisionTreeClassifier(criterion='entropy', random_state=0)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print("Predicted:", y_pred)
print("Actual   :", list(y_test.values))
print("Accuracy Score:", accuracy_score(y_test, y_pred) * 100)

# Optional: Visualize the tree
plt.figure(figsize=(12,8))
tree.plot_tree(model, filled=True, feature_names=['Subject', 'Hours_Studied', 'Tuition'], class_names=['Low', 'High'])
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import svm
import seaborn as sns; sns.set(font_scale=1.2)

x = pd.read_csv("/home/aditya/Desktop/Companies.csv")
x['P&L'] = np.random.choice(['Profit','Loss'], size=len(x))

x.head()

sns.lmplot(x='Revenue (Cr)', y='Employees', data=x, hue='P&L', palette='Set1', fit_reg=False, scatter_kws={"s":70});
type_label = np.where(x['P&L']=='Profit', 1, 0)
x_list = x.columns.values[1:6].tolist()
print(x_list)

parameter = x[['Revenue (Cr)','Employees']].values

model = svm.SVC(kernel='linear')
model.fit(parameter, type_label)

# Separating Hyperplane
w = model.coef_[0]
a = -w[0]/w[1]
xx = np.linspace(30, 60)
yy = a*xx - (model.intercept_[0])/w[1]

#Support Vectors
b = model.support_vectors_[0]
yy_down = a * xx + ( b[1] - a * b[0])
b = model.support_vectors_[-1]
yy_up = a * xx + ( b[1] - a * b[0])

sns.lmplot(x='Revenue (Cr)', y='Employees', data=x, hue='P&L', palette='Set1', fit_reg=False, scatter_kws={"s":70});
plt.plot(xx, yy, linewidth=2, color='black')
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
```
