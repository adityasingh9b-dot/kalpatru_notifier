```python
# Enabling Camera:
import cv2
video_cap = cv2.VideoCapture(0)  # Captures the video and stores it in the variable 'video_cap'

while True:  #infinite loop
    ret, video_data = video_cap.read() #'video_cap.read()' returns two data, first True or False that stores in 'ret' and 'video_data' stores video as numpy array data format
    cv2.imshow("Aditya's Project Camera", video_data) # Opens camera tab with the title heading in "", and with video_data as display on screen
    if cv2.waitKey(1) == ord('a'):    # listens waitKey() milliseconds , if you press key inside ord() function, the camera stops (loop exits) 
        break

video_cap.release() # Turn off camera 
```

```python
import cv2

cap = cv2.VideoCapture(0)

cv2.namedWindow("FullScreenGame", cv2.WND_PROP_FULLSCREEN)
cv2.setWindowProperty("FullScreenGame", cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    frame = cv2.flip(frame, 1)  # Mirror for natural feel
    
    cv2.imshow("FullScreenGame", frame)
    
    key = cv2.waitKey(1)
    if key == 27:  # ESC key to exit
        break

cap.release()
cv2.destroyAllWindows()
```
